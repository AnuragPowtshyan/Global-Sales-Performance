```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Global Sales Dataset.csv', encoding='latin-1')
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df['Postal Code'] = df['Postal Code'].fillna('No Data')
```

```python
df['Net Profit'] = df['Profit'] - df['Shipping Cost']
```

```python
df.to_csv('/content/cleaned_sales.csv', index=False)
```
